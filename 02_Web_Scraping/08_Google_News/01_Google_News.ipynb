```python
#Everything used in this code is only for learning purpose
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
chromeOptions = Options()
```

```python
def get_page(url,search_text):
    driver = webdriver.Chrome(options=chromeOptions)
    driver.get(url)
    time.sleep(3)
    search_ele = driver.find_element_by_xpath("//input[@role='combobox']")
    search_ele.clear()
    search_ele.send_keys(search_text)
    time.sleep(2)
    search_ele.send_keys(Keys.RETURN)
    time.sleep(6)
    driver.refresh();
    webpage = driver.page_source
    time.sleep(1)
    driver.quit()
    
    return webpage
```

```python
def print_data(containers):
    for i,a in enumerate(containers):
        print(i,10*"==")
        print("Heading: ",a.h3.text)
        try:
            print("Date: ",a.time['datetime'])
        except:
            print("Not available")
        print("Text :",a.span.text)
        print("\n")
```

```python
def get_df(containers):
    data = {}
    data['heading'] = []
    data['datetime'] = []
    data['text'] = []
    for a in containers:
        data['heading'].append(a.h3.text)
        try:
            data['datetime'].append(a.time['datetime'])
        except:
            data['datetime'].append("na")
        data['text'].append(a.span.text)
    df = pd.DataFrame(data)
    return df
```

```python
url = "https://news.google.com/"
search_text = "accident in texas on 10 july"
webpage = get_page(url,search_text)
```

```python
page_soup = bs(webpage,"html.parser")
containers = page_soup.findAll("article")
```

```python
print_data(containers)
df = get_df(containers)
```
